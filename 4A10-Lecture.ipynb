```python
# # 05/12/2025
# Conditional statements questions


# Strings


for i in range(6): # Rows
    for j in range(6): # Columns
        if i==0 or j==0 or i==6//2 or j==5:  # R LETTER
            print("*",end="")
        else:
            print(" ",end="")
    print(end="  ")
    for j in range(6):  # E LETTER
        if i==0 or j==0 or i==5 or i==6//2:
            print("*",end="")
        else:
            print(" ",end="")
    print(end="  ")
    for j in range(6):   # V LETTER
        if (i==j and j<6//2) or (j==6-i-1 and j>=6//2):
            print("*",end="")
        else:
            print(" ",end="")
    print(end="  ")
    for j in range(6):  # A LETTER
        if i==0 or j==0 or i==6//2 or j==5:
            print("*",end="")
        else:
            print(" ",end="")
    print(end="    ")
    for j in range(6):  # N LETTER
        if j==0 or j==5 or i==j:
            print("*",end="")
        else:
            print(" ",end="")
    print(end="    ")
    for j in range(6):   # T letter
        if i==0 or j==6//2:
            print("*",end="")
        else:
            print(" ",end="")
    print(end="     ")
    for j in range(6):   # H letter
        if j==0 or j==5 or i==6//2:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
# Fruit Ful Functions -- Functions with return type are called fruitful functions
# void functions - functions without any return type are called void functions


# Parameterized Constructor


class Car:
    def __init__(self,color,brand):
        self.colorofcar=color # (benz.colorofcar) (audi.colorofcar)
        self.brandofcar=brand # (benz.brandofcar) (audi.brandofcar)
    
    def colorprinting(s):
        print("Color of the car",s.colorofcar) #(benz.colorofcar)
    
    def brandprinting(f):
        print("Brand of the car",f.brandofcar)

benz=Car("Red","BENZ")
benz.colorprinting()
benz.brandprinting()

print()

audi=Car("black","audi")
audi.colorprinting()
audi.brandprinting()

print()


range_rover=Car("Marron","range_rover")
range_rover.colorprinting()
range_rover.brandprinting()
```

```python
# single inheritance

class Grandparent:
    def land(s):
        print("Land Property")

class Parent(Grandparent):
    def house(s):
        print("House")
a=Parent()
a.land()
a.house()
```

```python
# Multilevel
# gp->p->c
class Grandparent:
    def land(s):
        print("Land Property")

class Parent(Grandparent):
    def house(s):
        print("House")

class Child(Parent):
    def mansion(s):
        print("mansion house")
a=Child()
a.land()
a.house()
a.mansion()
```

```python
# # Abstraction  - hiding the method implementation
# or simply hiding the data 

# we use abc module for the abstraction
# decorator - it is used to specify some importance to the
# method(simple definition)
# here we use @abstractmethod as a decorator

# to specify the abstractmethods from general methods

# ABC - abstract base class

from abc import ABC, abstractmethod

class classname(ABC):
    @abstractmethod
    def abstractmethod1(self):
        pass
    @abstractmethod
    def abstractmethod2(self):
        pass
    @abstractmethod
    def abstractmethod3(self):
        pass
    def generalmethod(self):
        print("general method")
class Child(classname):
    pass
    # abstract methods implementation
```

```python
# Encapsulation

# public , private, protected

# public - def hello()
# private - def __hello() , __a
# protected - def _hello() , _a

class A:
    __a="hello"
    def hello(self):
        print("Public method")
    def __private(self):
        print("private method")
        
    def publicmethod(self):
        self.__private()
        print(self.__a)
class B(A):
    pass
    
a=B()
a.hello()
a._A__private()
a.publicmethod()
    
```
