```python
import pandas as pd

# A series is one dimensional data
# like a single column in excel

s=pd.Series([10,20,30])
print(s)

print()
d=pd.Series([10,20,30],index=['a','b','c'])
print(d)


print(d['a'])
```

```python
# A dataframe is a 2 dimensional table made of 
# rows , columns , row index, column names


import pandas as pd
data={
    "name":['A','B','C','D','E',"F"],
    "Age":[20,21,22,23,24,25]
}

df=pd.DataFrame(data,index=[1,2,3,4,5,6])
print(df)
print()
print(df.head())  # First 5 rows it will take

print(df.tail()) # last 5 rows 


# Shape of dataframe
#shows(rows,columns)
print()
print(df.shape)



# Access the columns

df[['name','Age']]
```
