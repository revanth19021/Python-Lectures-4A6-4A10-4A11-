```python
class car:
    def __init__(self,color,speed):
        self.colorofcar=color  # self=benz benz.colorofcar
        self.speedofcar=speed
    def colormethod(revanth):
        print(revanth.colorofcar)
    def speedmethod(prince):
        print(prince.speedofcar)

benz=car("black",250)
benz.colormethod()
benz.speedmethod()
print()
audi=car("grey",300)
audi.colormethod()
audi.speedmethod()
print()
nano=car("yellow",120)
nano.colormethod()
nano.speedmethod()
```
