```python
# SNAKE GAME USING PYGAME

Snake game rules (simple):
    1.Snake moves continuously
    2.Snake eats food
    3.Snake grows after eating food
    4.If snake hits wall â†’ game over
    5.If snake hits itself â†’ game over
    
    
We will build:
ðŸŸ© Snake â†’ rectangles
ðŸ”´ Food â†’ small rectangle
ðŸŽ® Arrow keys â†’ control snake
```

```python
import pygame
pygame.init()   #  starts pygame

screen=pygame.display.set_mode((400,300))  #opens a window 500 width, 500 height

x=100 # Starting position
running=True

while running:
    
    screen.fill((0,0,0)) # paint the screen
    
    for event in pygame.event.get():  # give me all the events that happened recently
        if event.type==pygame.QUIT:
            running=False
            # While the game is running, check all events.
            # If the user clicks the close button, stop the game
    x=x+1 # move right
    pygame.draw.rect(screen, (0,255,0), (x, 100, 20, 20))  # GREEN BOX
    
    pygame.display.update()
    
pygame.quit()

```
