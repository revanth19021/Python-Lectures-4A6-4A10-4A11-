```python
# Account class
class Account:
	def __init__(self, account_number, balance):
		self.account_number = account_number
		self.balance = balance
	
	def deposit(self, amount):
		self.balance += amount
		print(f"Balance after deposit: {self.balance}")
	
	def withdraw(self, amount):
		if amount <= self.balance:
			self.balance -= amount
			print(f"Balance after withdrawal: {self.balance}")
		else:
			print("Insufficient funds!")
	
	def display_balance(self):
		print(f"Current balance: {self.balance}")

# Customer class

class Customer:
	def __init__(self,name,account):
		self.name=name
		self.account=account
	
	def display_customer(self):
		print("Customer Details:")
		print(f"Customer Name: {self.name}")
		print(f"Account Number: {self.account.account_number}")
		print(f"Current balance: {self.account.balance}")

# Bank class
class Bank:
	def __init__(self, name,customer):
		self.name = name
		self.customer =customer 
	
	def set_customer(self, customer):
		self.customer = customer
	
	
	def display_customer(self):
		if self.customer:
			print(f"Customer in {self.name}: {self.customer.name}")
bank_name = input("Bank name: ") # Bank Input
customer_name = input("Customer name: ") # Name Input
account_number = input("Account number: ") # Account input
initial_balance = float(input("Initial balance: ")) # Intial Balance
account = Account(account_number, initial_balance)
customer = Customer(customer_name, account)
bank = Bank(bank_name,customer)
deposit_amount = float(input("Amount to Deposit: "))
account.deposit(deposit_amount)
bank.set_customer(customer)
withdraw_amount = float(input("Amount to Withdraw: "))
account.withdraw(withdraw_amount)
customer.display_customer()



```

```python
# Student class
class Student:
	def __init__(self, name, roll_number):
		self.name = name
		self.roll_number = roll_number
	def display_student(self):
		print("Student Details:")
		print(f"Student Name: {self.name}")
		print(f"Roll Number: {self.roll_number}")
	
	


# Teacher class
class Teacher:
	def __init__(self, name, subject):
		self.teachername = name
		self.subject = subject
	def display_teacher(self):
		print("Teacher Details:")
		print(f"Teacher Name: {self.teachername}")
		print(f"Subject: {self.subject}")


# Course class
class Course:
	def __init__(self, course_name, teacher, student):
		self.course_name = course_name
		self.teacher = teacher
		self.student = student
	def display_course(self):
		print("Course Details:")
		print(f"Course Name: {self.course_name}")
		print(f"Teacher: {self.teacher.teachername}")
		print(f"Student: {self.student.name}")
	
#Type your code here...	
student_name = input("Student name: ")
student_roll_number = input("Student roll number: ")
teacher_name = input("Teacher name: ")
subject = input("Subject: ")
course_name = input("Course name: ")


student = Student(student_name, student_roll_number)
teacher = Teacher(teacher_name, subject)
course = Course(course_name, teacher, student)


student.display_student()
teacher.display_teacher()
course.display_course()
```
