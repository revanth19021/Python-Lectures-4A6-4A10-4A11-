```python
#  A program that converts temperatures from Fahrenheit to Celsius and vice versa

print("1: fahrenheit  to celsius")
print("2: celcius to fahrenheit")


choice= input("Enter Choice: ")

if choice=='1':
    f=float(input("Enter the Temperature :  "))
    c=(f-32)*5/9
    print(f"{f}F: {c:.2f}C")
elif choice=='2':
    c=float(input("Enter the celsius: "))
    f=(c*9/5)+32
    print(f"{c}C: {f:.2f}F")
else:
    print("Invalid Choice")
```

```python
# Random Password Generator
import random
characters="abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()?"
password=""


for _ in range(9):
    password+=random.choice(characters)

print(password)



# One more way

import random
length=int(input("Enter the Password Length: "))

password=""

for i in range(length):
    password+=chr(random.randint(33,126))
print(password)

print(chr(65)) # Used to Convert into Character
```

```python
# Leap Year

year=int(input("Enter the year: "))

if((year%4==0 and year%100!=0) or (year%400==0)):
    print("Leap year")
else:
    print("Not leap year")
```

```python
text=input("Enter a string: ")

if text==text[::-1]:
    print("palindrome")
else:
    print("not palindrome")
    

    
    
# Second way

text=input("Enter the String: ")

rev=""

for ch in text:
    rev=ch+rev
print(rev)

if rev==text:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# A program that sorts a list of numbers in ascending or descending order.

numbers=list(map(int,input("Enter the list numbers: ").split()))
print(numbers)

order=input("enter a for ascending d for descending ").lower()

if order=='a':
    numbers.sort()
elif order=='d':
    numbers.sort(reverse=True)
else:
    print("Invalid Choice")
print(numbers)
```

```python
# Factorial Calculator

fact=1

n=int(input("Enter the number:  "))
for i in range(1,n+1):
    fact=fact*i
print(fact)
```

```python
# Base Conversion
def converting(num,base):
    digits="0123456789ABCDEF"
    res=""
    if num==0:
        return "0"
    while num>0:
        remainder=num%base
        res=digits[remainder]+res
        num=num//base
    return res
a=input()
b=int(input())
c=int(input())
d=int(a,b)

l=converting(d,c)
print(l)
```
