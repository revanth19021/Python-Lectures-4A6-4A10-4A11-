```python
# Line Chart
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Line Chart Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

# Best for: Sales over time, temperature changes, growth trends.
```

```python
# Bar Chart 

# Used for comparing categories.
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

# Best for: Comparing students, products, departments.
```

```python
# Histogram

# Used for distribution of data.
data = [10, 12, 15, 20, 22, 25, 30, 35, 40, 45]

plt.hist(data, bins=5)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Best for: Exam scores distribution, age distribution.
```

```python
# Scatter Plot

# Used to show relationship between two variables.

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

# Best for: Studying correlation (e.g., study hours vs marks).
```

```python
# Pie Chart

# Used to show percentages.

sizes = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
# Best for: Market share, budget distribution.
```

```python
# Box Plot

# Used to show spread and outliers

data = [5, 7, 8, 9, 10, 12, 15, 20, 30]

plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()

# Best for: Finding median, quartiles, outliers.
```

```python
| Chart Type | Used For                       |
| ---------- | ------------------------------ |
| Line       | Trends over time               |
| Bar        | Comparing categories           |
| Histogram  | Data distribution              |
| Scatter    | Relationship between variables |
| Pie        | Percentage share               |
| Box        | Spread & outliers              |
```
