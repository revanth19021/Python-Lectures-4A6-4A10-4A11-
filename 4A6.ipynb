```python
from abc import ABC,abstractmethod
class BankingApplication(ABC):
    @abstractmethod
    def createaccount():
        pass
class Revanth(BankingApplication):
    def __init__(self,name,balance,phonenumber,status):
        self.name=name
        self.balance=balance
        self.phonenumber=phonenumber
        self.status=status
    def createaccount(self):
        print(self.name,self.balance,self.phonenumber,self.status)
a=Revanth("revanth",30000,62815,True)
a.createaccount()
```
