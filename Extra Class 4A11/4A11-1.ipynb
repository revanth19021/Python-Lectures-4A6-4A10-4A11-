```python
# usage of super keyword in case of methods
class parent:
    def land(self,case):
        print("Vadodara Land 30 Acres",case)

class child(parent):
    def land(self,case):
        super().land("land litigation")
        print("Gujarat land 50 acres",case)

a=child()
a.land("Government Property")
```

```python
# Polymorphism
many forms of objects is called polymorphism
1. method overloading
2. method overriding
3. operator overloading
```

```python
# method overloading
# Same method name with different parameters

class A:
    def method1(self,a):
        print(a)
    def method1(self,b,c):
        print(b,c)
    def method1(self,b,c,d):
        print(b,c,d)
a=A()
# a.method1(3)
# a.method1(3,4)
a.method1(3,4,5)
```

```python
# indirect way of method overloading
1. by using variable length arguments
2. by using default arguments
```

```python
class A:
    def method1(self,*a):
        print(a)
a=A()
a.method1(3)
a.method1(3,4)
a.method1(3,4,5)
```

```python
# by using default arguments
# method overloading
# Same method name with different parameters

class A:
    def method1(self,b="hello",c="revanth"):
        print(b,c)
a=A()
a.method1()
a.method1(4)
```

```python
# method overriding
# 1. with the help of super() keyword we can implement method overriding
class company:
    def company(self):
        print("CodeTantra")
    def idcard(self,branch):
        print("id card of general",branch)

class intern(company):
    def stiphend(self,money):
        print("intern salary is",money)
    def idcard(self,branch):
        super().idcard("developer")
        print("id card of intern",branch)

class Employee(company):
    def salary(self,money):
        print("Employee salary is",money)
    def idcard(self,branch):
        super().idcard("trainer")
        print("Id card of employee",branch)

revanth=Employee()
revanth.salary(10000)
revanth.idcard("trainer")
revanth.company()
print()
rahul=intern()
rahul.stiphend(100)
rahul.idcard("developer")
rahul.company()
```

```python
# 2. with the help of parent class instance or parent class name we can implement method overriding
class parent:
    def method1(self,a,b):
        print("parent method",a,b)
class child(parent):
    def method1(self,a,b):
        parent().method1("revanth","hello")
        print("child method",a,b)
a=child()
a.method1(1,2)
```
