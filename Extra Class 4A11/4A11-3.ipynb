```python
# Inheritance -> Super() usage
it is used to call the parent class properties from the child class
if both parent and child consists of same properties
```

```python
# usage of super keyword in case of methods
class parent:
    def land(self,case,property_share):
        print("Vadodara Land 30 Acres",case,property_share)

class child(parent):
    def land(self,case):
        super().land("land litigation",2)
        print("Gujarat land 50 acres",case)

a=child()
a.land("Government Property")
```

```python
# usage of super keyword incase of constructors and variables

class parent:
    def __init__(self):
        pass
    def methodparent(s):
        print("Parent method")

class child(parent):
    def __init__(s,a,c):
        super().__init__()
        s.a=a
        s.c=c
    def methodchild(revanth):
        print(revanth.c,revanth.a)
a=child(5,6)
a.methodparent()
a.methodchild()
```
